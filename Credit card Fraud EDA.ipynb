How many transactions are fraudulent, and how many are legitimate?

```python
fraud_count = df[df['Class'] == 1].shape[0]
legitimate_count = df[df['Class'] == 0].shape[0]

print(f"Fraudulent transactions: {fraud_count}")
print(f"Legitimate transactions: {legitimate_count}")
```

What percentage of transactions are fraudulent?

```python
fraud_percentage = df['Class'].value_counts(normalize=True)[1] * 100
print(f"Percentage of fraudulent transactions: {fraud_percentage:.2f}%")
```

What are the minimum, maximum, mean, and median values for
numerical columns like Amount?

```python
amount_min = df['Amount'].min()
amount_max = df['Amount'].max()
amount_mean = df['Amount'].mean()
amount_median = df['Amount'].median()

print(f"Minimum value of Amount: {amount_min}")
print(f"Maximum value of Amount: {amount_max}")
print(f"Mean value of Amount: {amount_mean:.2f}")
print(f"Median value of Amount: {amount_median}")
```

What is the maximum transaction amount in the dataset, and is it
fraudulent?

```python
max_amount_row = df.loc[df['Amount'].idxmax()]

max_amount = max_amount_row['Amount']
is_fraudulent = max_amount_row['Class']

print(f"Maximum transaction amount: {max_amount}")
print(f"Is the transaction fraudulent? {'Yes' if is_fraudulent == 1 else 'No'}")
```

Can we create a bar chart showing the count of fraudulent vs.
legitimate transactions?

```python
transaction_counts = df['Class'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(transaction_counts.index, transaction_counts.values, color=['blue', 'red'])

plt.xticks([0, 1], ['Legitimate', 'Fraudulent'])
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Fraudulent vs Legitimate Transactions')


plt.show()
```

What does the histogram of transaction amounts look like?

```python

fraudulent = df[df['Class'] == 1]['Amount']
legitimate = df[df['Class'] == 0]['Amount']

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

axs[0].hist(fraudulent, bins=50, alpha=0.7, color='red')
axs[0].set_xlabel('Transaction Amount')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Histogram of Fraudulent Transactions')

axs[1].hist(legitimate, bins=50, alpha=0.7, color='green')
axs[1].set_xlabel('Transaction Amount')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Histogram of Legitimate Transactions')

plt.tight_layout()
plt.show()
```

Conclusion:

Left-skewed distributions for both fraudulent and legitimate transactions indicate that most of the transactions are on the larger side, with only a few smaller transactions contributing to the skew.

This skewness doesn't necessarily point to an issue with the data; itâ€™s a typical pattern in many financial datasets where smaller amounts are more common

Can we use a heatmap to visualize the correlation between numerical
features?

```python

correlation_matrix = df.corr()


correlation_matrix_rounded = correlation_matrix.round(1)


plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix_rounded, annot=True, cmap='coolwarm', fmt='.1f', linewidths=1, linecolor='black', cbar_kws={'shrink': 0.8})

plt.title('Correlation Heatmap of Numerical Features')

plt.show()

```

Summary of Correlation Coefficients:

Positive Correlation (+1 to 0):

- +1: Perfect positive correlation (perfectly aligned increases).
- 0.5 to 0.9: Moderate positive correlation.
- 0.1 to 0.4: Weak positive correlation.
- 0: No correlation.

Negative Correlation (-1 to 0):

- -1: Perfect negative correlation (perfectly opposite changes).
- -0.5 to -0.9: Moderate negative correlation.
- -0.1 to -0.4: Weak negative correlation.
- 0: No correlation.
- Constant Correlation:

If the correlation coefficient between two variables is exactly 1 or -1 (perfect correlation), it means that one variable perfectly determines the other, either directly or inversely.
In practice, this is rare for most real-world datasets, except for perfectly linear relationships.
